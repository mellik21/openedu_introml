```python
import numpy as np
import pandas as pd
train_data = pd.read_csv("report.csv", delimiter=',')
```

```python
train_data.head()
```

```python
X = train_data.drop('TARGET', axis=1)
```

```python
y = train_data.TARGET
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
reg = LogisticRegression(random_state=2019, solver='lbfgs')
```

```python
reg.fit(X, y.values.ravel())
```

```python
reg.predict_proba([[0.363, 0.178, 0.43, 0.292, 0.575, 0.872, 0.023, 0.161]])
```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=5, p=2)
neigh.fit(X, y)
```

```python
neigh.predict([[0.363, 0.178, 0.43, 0.292, 0.575, 0.872, 0.023, 0.161]])
```

```python
neigh.predict_proba([[0.363, 0.178, 0.43, 0.292, 0.575, 0.872, 0.023, 0.161]])
```

```python
neigh.kneighbors([[0.363, 0.178, 0.43, 0.292, 0.575, 0.872, 0.023, 0.161]])
```
