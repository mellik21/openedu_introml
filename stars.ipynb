```python
import numpy as np
import pandas as pd
train_data = pd.read_csv("report.csv", delimiter=',')
```

```python
train_data.head()
```

```python
train_data.MIP.mean()
```

```python
X = train_data.drop('TARGET', axis=1)
```

```python
y = train_data.TARGET
```

```python
X.describe()
```

```python
mip = []
for var in X.MIP :
    res = (var - 77.492188) / (96.414062 - 77.492188) 
    mip.append(res)
X.MIP = mip
```

```python
X.MIP.mean()
```

```python
X.STDIP = (X.STDIP-31.006264)/(83.796909 - 31.006264)
X.EKIP = (X.EKIP -0.190183 )/(2.503509 -0.190183 )
X.SIP = (X.SIP + 0.632872)/(11.832445 +0.632872 )
X.MC = (X.MC -0.414716 )/(170.805184 -0.414716 )
X.STDC = (X.STDC -8.425768 )/(98.321744 - 8.425768)
X.EKC = (X.EKC + 1.227876 )/(25.968793 + -1.227876)
X.SC = (X.SC + 1.837085 )/(733.935093 + 1.837085 )
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
reg = LogisticRegression(random_state=2019, solver='lbfgs')
```

```python
reg.fit(X, y.values.ravel())
```

```python
reg.predict_proba([[0.363, 0.178, 0.43, 0.292, 0.575, 0.872, 0.023, 0.161]])
```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=5, p=2)
neigh.fit(X, y)
```

```python
neigh.predict([[0.363, 0.178, 0.43, 0.292, 0.575, 0.872, 0.023, 0.161]])
```
