```python
import numpy as np
import pandas as pd
DATA = pd.read_csv("kmeans.csv", delimiter=',', index_col='Object')
```

```python
coords = DATA.drop('Cluster', axis=1)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3, init=np.array([[11.0, 10.33], [11.86, 12.71], [7.2, 12.8]]), max_iter=100, n_init=1)
```

```python
model = kmeans.fit(coords)
```

```python
model.labels_.tolist()
```

```python
alldistances = kmeans.fit_transform(coords)
```

```python
DATA['labels'] = model.labels_.tolist()
```

```python
DATA['distances'] = alldistances.tolist()
```
